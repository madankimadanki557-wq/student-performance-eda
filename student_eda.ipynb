```python
import pandas as pd

data = {
    "Name": ["Ravi", "Sita", "John", "Priya", "Arun"],
    "Math": [80, 92, 75, 88, 95],
    "Science": [85, 90, 70, 92, 98]
}

df = pd.DataFrame(data)

df
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df[["Math", "Science"]].mean().plot(kind="bar")

plt.show()
```

```python
print("EDA Summary")
print("Average Math Score:", df["Math"].mean())
print("Average Science Score:", df["Science"].mean())
```

```python
print("Maximum Math Score:", df["Math"].max())
print("Minimum Math Score:", df["Math"].min())
```

```python
df["Total"] = df["Math"] + df["Science"]
df
```

```python
top_student = df.loc[df["Total"].idxmax()]
print(top_student)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Total"])
plt.title("Student Total Scores")
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.show()
```
